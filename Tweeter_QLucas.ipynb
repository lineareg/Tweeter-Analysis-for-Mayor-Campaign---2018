# T3. We can use a different vectorizer to apply weights based on the frequency of the term in a Document

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# now apply weights based on the inverse document frequency
tfidf3 = TfidfVectorizer(use_idf=True, lowercase=False,stop_words='english',ngram_range=(1, 2),max_df = 0.5, min_df= .08) #define the transformation
tf3_dm = tfidf3.fit_transform(tweet_frame['tweetText']) #apply the transformation
print(tf3_dm.shape)
pd.DataFrame(tf3_dm.toarray(),columns = tfidf3.get_feature_names()).head(10)
```

```python
names = tfidf3.get_feature_names()   #create list of feature names
count = np.sum(tf3_dm.toarray(), axis = 0) # add up feature counts 
count2 = count.tolist()  # convert numpy array to list
count_df = pd.DataFrame(count2, index = names, columns = ['count']) # create a dataframe from the list
count_df.sort_values(['count'], ascending = False)  #arrange by count instead
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# now apply weights based on the inverse document frequency
tfidf4 = TfidfVectorizer(smooth_idf=True, lowercase=True,stop_words='english',max_features=20,ngram_range=(1, 2)) #define the transformation
tf4_dm = tfidf4.fit_transform(tweet_frame['tweetText']) #apply the transformation
print(tf4_dm.shape)
pd.DataFrame(tf4_dm.toarray(),columns = tfidf4.get_feature_names()).head(10)
```

```python
names = tfidf4.get_feature_names()   #create list of feature names
count = np.sum(tf4_dm.toarray(), axis = 0) # add up feature counts 
count2 = count.tolist()  # convert numpy array to list
count_df = pd.DataFrame(count2, index = names, columns = ['count']) # create a dataframe from the list
count_df.sort_values(['count'], ascending = False)  #arrange by count instead
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# now apply weights based on the inverse document frequency
tfidf5 = TfidfVectorizer(sublinear_tf=True,norm="l2",lowercase=False,stop_words='english',max_features=20,ngram_range=(1, 2)) #define the transformation
tf5_dm = tfidf5.fit_transform(tweet_frame['tweetText']) #apply the transformation
print(tf5_dm.shape)
pd.DataFrame(tf5_dm.toarray(),columns = tfidf5.get_feature_names()).head(10)
```

```python
names = tfidf5.get_feature_names()   #create list of feature names
count = np.sum(tf5_dm.toarray(), axis = 0) # add up feature counts 
count2 = count.tolist()  # convert numpy array to list
count_df = pd.DataFrame(count2, index = names, columns = ['count']) # create a dataframe from the list
count_df.sort_values(['count'], ascending = False)  #arrange by count instead
```
